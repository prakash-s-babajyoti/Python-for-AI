### The "Why"
In the real world, things go wrong. A user might type "five" instead of "5", or your program might try to open a file that doesn't exist. If you don't handle these situations, your program will "crash" (stop running). **Error Handling** lets you catch these mistakes and handle them gracefully. 

Additionally, most programs need to save data permanently. **File I/O** (Input/Output) allows you to read from and write to files like `.txt`, `.csv`, and `.json`.

### 1. Error Handling: Try / Except
You wrap code that *might* fail in a `try` block. If an error occurs, the `except` block catches it.

```
try:
    number = int(input("Enter a number: "))
    result = 10 / number
    print(f"Result: {result}")
except ValueError:
    print("Error: Please enter a valid whole number.")
except ZeroDivisionError:
    print("Error: You cannot divide by zero!")
finally:
    print("This code runs no matter what happened.")
```

### 2. File I/O: The 'With' Statement
When working with files, it's critical to close them after you're done. The `with` statement (called a **Context Manager**) does this for you automatically.

```
# Writing to a file
with open("note.txt", "w") as file:
    file.write("Hello, World!")

# Reading from a file
with open("note.txt", "r") as file:
    content = file.read()
    print(content)
```

### 3. Handling Structured Data: JSON
Most modern applications share data using JSON (JavaScript Object Notation). Python has a built-in `json` module to handle this.

```
import json

user_data = {"name": "Alice", "score": 100}

# Convert dictionary to JSON string
json_string = json.dumps(user_data)

# Save to a file
with open("user.json", "w") as f:
    json.dump(user_data, f)
```

### ⚠️ Common Pitfalls
* **Broad Except Blocks:** Never use `except: pass` or `except Exception:`. This hides errors that you *should* see. Always catch specific errors like `FileNotFoundError`.
* **Forgetting to Close Files:** If you don't use `with`, and your program crashes before `file.close()`, you might lose data or lock the file.

## 💻 Hands-On Examples

**Example 1: Safe Division Function**

```
def safe_divide(a, b):
    try:
        return a / b
    except (TypeError, ZeroDivisionError) as e:
        return f"Invalid calculation: {e}"

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "2"))
```

**Example 2: Appending to a Log File**

```
import datetime

def log_message(msg):
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open("activity.log", "a") as f:
        f.write(f"[{timestamp}] {msg}\n")

log_message("User logged in")
log_message("Data saved successfully")
```
