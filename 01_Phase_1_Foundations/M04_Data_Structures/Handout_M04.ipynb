## 📖 The Handout
### The "Why"
So far, you've stored single pieces of data in variables (like `score = 100`). But what if you have 1,000 scores? Or what if you want to store a user's name, email, and password together? Python provides **Data Structures**—special containers that hold multiple pieces of data. 

### The Core Four
1. **Lists `[]`**: Ordered, changeable. Used for collections of items.
2. **Dictionaries `{}`**: Key-Value pairs. Used for labeled data (like JSON).
3. **Tuples `()`**: Ordered, *unchangeable* (immutable). Used for fixed data.
4. **Sets `{}`**: Unordered, *unique* items only. Used for fast math/comparisons.

### 1. Lists: Slicing & Unpacking
Lists are the most common data structure. You can slice them to get subsections, or unpack them into variables.

```
fruits = ["apple", "banana", "cherry", "date"]

# Indexing (starts at 0)
print(fruits[0]) # "apple"

# Slicing [start:end] (end is excluded)
print(fruits[1:3]) # ["banana", "cherry"]

# Unpacking
first, second, *rest = fruits
print(rest) # ["cherry", "date"]
```

### 2. Dictionaries: Key-Value Pairs
Dictionaries don't use indexes (like 0, 1, 2). They use **keys** to look up **values**.

```
user = {
    "name": "Alice",
    "age": 25,
    "is_active": True
}
print(user["name"]) # "Alice"
user["email"] = "alice@example.com" # Adding a new key
```

### 3. Built-in Superpowers: Enumerate & Zip
Python has built-in functions to make looping over data structures easier.
* `enumerate()` gives you both the index and the item.
* `zip()` lets you loop over two lists at the same time.

### 4. Comprehensions (The Pythonic Way)
Comprehensions let you create a new list or dictionary from an old one in a single line of code!

```
numbers = [1, 2, 3, 4]
# Make a new list with doubled numbers:
doubled = [n * 2 for n in numbers]
```

### ⚠️ Common Pitfalls
* **IndexError:** Trying to access `fruits[10]` when the list only has 4 items will crash your code.
* **KeyError:** Trying to access `user["phone"]` if the dictionary doesn't have a "phone" key. Use `user.get("phone")` instead—it returns `None` safely.

## 💻 Hands-On Examples

**Example 1: The Power of Sets**

```
# Sets automatically remove duplicates!
messy_data = ["apple", "apple", "banana", "banana"]
clean_data = list(set(messy_data))
print(clean_data) # ['apple', 'banana']
```

**Example 2: Zip and Enumerate**

```
names = ["Alice", "Bob"]
scores = [90, 85]

# Zip combines them:
for name, score in zip(names, scores):
    print(name, "got a", score)

# Enumerate numbers them:
for index, name in enumerate(names):
    print("Student", index, "is", name)
```
