## 📖 The Handout
### The "Why"
As your programs get larger, writing the same code over and over becomes exhausting and prone to errors. **Functions** allow you to package a block of code, give it a name, and reuse it whenever you want. Think of a function as a mini-machine: you put raw materials in (arguments), the machine does some work, and it spits a finished product out (return value).

### The Syntax: Defining and Calling
You define a function using the `def` keyword, followed by the name and parentheses.

```
# 1. Defining the function
def greet_user(name):
    greeting = "Hello " + name + "!"
    return greeting

# 2. Calling the function
message = greet_user("Alice")
print(message)
```

### Arguments, kwargs, and Defaults
You can make your functions highly flexible by providing default values or allowing a variable number of arguments.

```
# Default parameter: If you don't provide a status, it defaults to "Standard"
def create_profile(username, status="Standard"):
    return "User: " + username + " | Status: " + status

print(create_profile("JohnDoe"))           # Uses default
print(create_profile("Admin99", "Admin"))  # Overrides default
```

### The Concept of Scope
Variables created *inside* a function cannot be accessed *outside* of it. This is called **Local Scope**.

```
def calculate_tax(amount):
    tax_rate = 0.20 # This is a LOCAL variable
    return amount * tax_rate

# print(tax_rate) -> ERROR! Python doesn't know what tax_rate is outside the function.
```

### ⚠️ Common Pitfalls
* **Forgetting to `return`:** If your function just uses `print()`, it doesn't give the data back to the rest of your program. If you assign a printed function to a variable, the variable will be `None`. Always `return` your final answer!
* **Calling a function before it's defined:** Python reads top to bottom. You cannot call `my_function()` on line 5 if you don't define it until line 10.

*Run these examples in your code editor to see what happens!*

**Example 1: Multiple Parameters**

```
def calculate_area(width, height):
    area = width * height
    return area

living_room = calculate_area(15, 20)
print("The living room is", living_room, "square feet.")
```

**Example 2: A Lambda Function (One-Liner)**
Lambda functions are quick, anonymous functions used for simple operations.

```
# A normal function
def double(x):
    return x * 2

# The exact same function as a Lambda
double_lambda = lambda x: x * 2

print(double(5))
print(double_lambda(5))
```
