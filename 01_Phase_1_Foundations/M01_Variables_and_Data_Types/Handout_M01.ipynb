## 📖 The Handout
### The "Why"
In Python, programs need to remember information to do anything useful. Whether it's a user's name, their bank balance, or a list of their friends, we store this information in **variables**. Think of a variable as a labeled box where you can put data. 

### The Syntax
To create a variable in Python, you just invent a name, use the equals sign `=`, and assign it a value. You don't need to declare what *type* of data it is—Python figures it out automatically!

```python
# Storing a string (text)
player_name = "Mario"

# Storing an integer (whole number)
lives_remaining = 3

# Storing a float (decimal number)
current_health = 85.5

# Storing a boolean (True/False)
is_game_over = False
```

### ⚠️ Common Pitfalls
* **Typos in variable names:** If you create `player_name` but later try to print `Player_name` (capital P), Python will crash. Variables are case-sensitive.
* **Using spaces in names:** `player name = "Mario"` will throw an error. Always use underscores `_` instead of spaces (this is called *snake_case*).
