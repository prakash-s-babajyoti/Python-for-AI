## 📖 The Handout
### The "Why"
So far, our code runs in a straight line from top to bottom. But what if we only want to run code *sometimes* (like granting access if a password is correct)? Or what if we want to run the same code *multiple times* (like emailing 100 customers)? We use **Control Flow** (`if/else` statements and `loops`) to control the direction our program takes.

### The Syntax: If / Elif / Else
The `if` statement checks a condition. If it's `True`, the indented code runs.

```
age = 18

if age >= 18:
    print("You can vote!")
elif age == 17:
    print("You can vote next year!")
else:
    print("You are too young to vote.")
```

### The Syntax: Loops
**For Loops:** Used when you know exactly how many times to repeat something (or when iterating over a collection).

```
# Range generates numbers from 0 up to (but not including) 5
for i in range(5):
    print("Hello", i)
```

**While Loops:** Used when you want to repeat something *until* a condition changes.

```
countdown = 3
while countdown > 0:
    print(countdown)
    countdown = countdown - 1
print("Blastoff!")
```

### ⚠️ Common Pitfalls
* **Indentation Errors:** Python uses spaces (indentation) to know which code belongs inside an `if` statement or a loop. If you forget to indent, Python throws an `IndentationError`.
* **Infinite Loops:** If you create a `while` loop but forget to update the condition (like forgetting `countdown = countdown - 1`), the loop will run forever and crash your program!

## 💻 Hands-On Examples
*Run these examples in your code editor to see what happens!*

**Example 1: The Password Checker**

```
password_attempt = "12345"
actual_password = "password123"

if password_attempt == actual_password:
    print("Access Granted")
else:
    print("Access Denied")
```

**Example 2: Skipping with 'continue' and stopping with 'break'**

```
for num in range(1, 6):
    if num == 3:
        continue # Skips the number 3
    if num == 5:
        break # Stops the loop entirely when it hits 5
    print(num)
```
