### The "Why"
So far, you've used strings, integers, and lists. But what if you are building a video game and need to represent a "Player"? A player isn't just a string; they have data (health, name, inventory) and behaviors (attack, jump, heal). 
**Object-Oriented Programming (OOP)** allows you to create your own custom data types (Classes) to model real-world concepts.

### 1. Classes & Objects
A **Class** is a blueprint. An **Object** is the actual house built from that blueprint.

```
# Defining the Blueprint (Class)
class Dog:
    pass

# Creating Objects (Instances)
my_dog = Dog()
your_dog = Dog()
```

### 2. Constructors (`__init__`) and Instance Variables
To give an object data when it's created, we use a special "dunder" (double underscore) method called `__init__`. The word `self` is used to attach data to the specific object being created.

```
class Dog:
    # The Constructor
    def __init__(self, name, age):
        self.name = name  # Instance variable
        self.age = age    # Instance variable

    # A Method (Behavior)
    def bark(self):
        return self.name + " says woof!"

# Creating an object with data
my_dog = Dog("Buddy", 3)
print(my_dog.bark()) # "Buddy says woof!"
```

### 3. Instance vs Class Variables
* **Instance Variables:** Data unique to each object (like `self.name`).
* **Class Variables:** Data shared across *all* objects of that class.

```
class Dog:
    species = "Canis familiaris" # Class variable (all dogs share this)

    def __init__(self, name):
        self.name = name # Instance variable (unique to each dog)
```

### 4. Inheritance & Polymorphism
**Inheritance** allows a new class to inherit properties from an existing class. **Polymorphism** allows child classes to override behaviors.

```
class Animal:
    def speak(self):
        return "Some sound"

class Cat(Animal): # Cat inherits from Animal
    def speak(self): # Polymorphism: Overriding the speak method
        return "Meow"

my_cat = Cat()
print(my_cat.speak()) # Prints "Meow"
```

### 5. Magic / Dunder Methods
Python uses special methods like `__init__`, `__str__` (for printing), and `__len__` (for length) to add "magic" behaviors to your classes.

### ⚠️ Common Pitfalls
* **Forgetting `self`:** When defining a method inside a class, the first parameter *must* be `self`. If you define `def bark():` instead of `def bark(self):`, Python will throw a TypeError.
* **Over-engineering:** Don't use a class if a simple dictionary or function will do the job. Only use classes when you need to bundle data and behavior together.

## 💻 Hands-On Examples

**Example 1: The Magic of `__str__`**

```
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        
    def __str__(self):
        return self.title + " by " + self.author

my_book = Book("1984", "George Orwell")
# Without __str__, this would print a memory address like <__main__.Book object at 0x...>
# With __str__, it prints beautifully:
print(my_book) 
```
